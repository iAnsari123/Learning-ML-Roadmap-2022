```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.rename(
    columns={
        "sepal_length": "sepal_length_lomba",
        "sepal_width": "sepal_width_mota",
        "petal_length": "petal_length_lomba",
        "petal_width": "petal_width_mota",
        "species": "species_jat",
    },
    inplace=True,
)
df.head()
```

```python
from statistics import mode

for i in df["species_jat"].value_counts().index:
    print(f"The Central tendency matrix for {i}: ")
    print(
        df[df["species_jat"] == i][
            [
                "sepal_length_lomba",
                "sepal_width_mota",
                "petal_length_lomba",
                "petal_width_mota",
            ]
        ]
        .agg(["mean", "median", mode])
        .T
    )

    print("\n")
```

```python
range_of_species = df["species_jat"].value_counts().index
for i in range_of_species:
    print(f"The range of {i} is: ")
    print(
        df[df["species_jat"] == i][
            [
                "sepal_length_lomba",
                "sepal_width_mota",
                "petal_length_lomba",
                "petal_width_mota",
            ]
        ]
        .agg(["min", "max", "std", "var"])
        .T
    )
    print("\n")
```
