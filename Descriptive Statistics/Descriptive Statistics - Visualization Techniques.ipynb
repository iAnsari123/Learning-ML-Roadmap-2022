### Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.info()
```

## Univartiante Analysis

### 1. Working with qualitative/categorical data

##### What is a frequency distribution?
The frequency of a value is the number of times it occurs in a dataset. A frequency distribution is the pattern of frequencies of a variable. Itâ€™s the number of times each possible value of a variable occurs in a dataset.


<img src="https://www.scribbr.com/wp-content/uploads/2022/07/02.png" height="400" width= "600">

```python
df["class"].value_counts().plot(kind="barh")
```

```python
df["class"].value_counts().plot(kind="pie", autopct="%.2f")
```

```python
df["class"].value_counts(normalize=True).cumsum().plot(kind="line", marker="o")
```

### 2. Working with quantative/numerical data

```python
sns.kdeplot(df["age"], fill=True)
```

```python
df["age"].plot(kind="hist", bins=30)
```

```python
sns.boxplot(df["age"], orient="h")
```

## Bivariate Analysis

### 1. Working with categorical vs numerical plots

```python
sns.boxplot(x="class", y="age", data=df)
```

```python
sns.barplot(x="class", y="age", data=df)
```

### 2. Working with numerical vs numerical plots

```python
sns.scatterplot(x="age", y="fare", data=df)
plt.yticks(np.arange(0, 501, 20), [i for i in range(0, 501, 20)])
plt.show()
```

### 3. Working with categorical vs categorical plots

```python
pd.crosstab(df["class"], df["survived"]).plot(kind="bar")
plt.legend(["Dead", "Survived"])
```
